```python
import os
import pandas as pd
```

```python
def filter(airport):
#     airport = airport.loc[(airport['Destination Airport']=="SFO")|(airport['Destination Airport']=="LAX")|
#                          (airport['Destination Airport']=="DEN")|(airport['Destination Airport']=="LAS")|
#                          (airport['Destination Airport']=="PDX")|(airport['Destination Airport']=="SEA")|
#                          (airport['Destination Airport']=="DFW")|(airport['Destination Airport']=="MCO")|
#                          (airport['Destination Airport']=="ORD")|(airport['Destination Airport']=="JFK")|
#                          (airport['Destination Airport']=="ATL"),["Carrier Code","Date (MM/DD/YYYY)","Flight Number",
#                                                                   "Tail Number","Destination Airport",
#                                                                   "Scheduled departure time",
#                                                                   "Actual departure time",
#                                                                   "Scheduled elapsed time (Minutes)",
#                                                                   "Actual elapsed time (Minutes)",
#                                                                   "Departure delay (Minutes)",
#                                                                   "Wheels-off time",
#                                                                   "Taxi-Out time (Minutes)",
#                                                                   "Delay Carrier (Minutes)",
#                                                                   "Delay Weather (Minutes)",
#                                                                   "Delay National Aviation System (Minutes)",
#                                                                   "Delay Security (Minutes)",
#                                                                   "Delay Late Aircraft Arrival (Minutes)",
#                                                                   "Origin Airport"]]
    airport = airport.rename(columns={'Destination Airport': 'Destination_Airport'})
    Destination_booleans = []
    for place in airport['Destination Airport']:
        if place == "SFO":
            Destination_booleans.append(True)
        elif place == "LAX":
            Destination_booleans.append(True)
        elif place == "DEN":
            Destination_booleans.append(True)
        elif place == "LAS":
            Destination_booleans.append(True)
        elif place == "PDX":
            Destination_booleans.append(True)
        elif place == "SEA":
            Destination_booleans.append(True)
        elif place == "DFW":
            Destination_booleans.append(True)
        elif place == "MCO":
            Destination_booleans.append(True)
        elif place == "ORD":
            Destination_booleans.append(True)
        elif place == "JFK":
            Destination_booleans.append(True)
        elif place == "ATL":
            Destination_booleans.append(True)
        else:
            Destination_booleans.append(False)
    is_long = pd.Series(Destination_booleans)
    airport = airport[is_long]

    return (airport)
    
    
#     
```

```python
ATL1 = pd.read_csv("Resources/CSV/American/ATL.csv", skiprows=7)[:-2]
ATL1["Origin Airport"] = "ATL"

# filter(ATL1)

DEN1 = pd.read_csv("Resources/CSV/American/DEN.csv", skiprows=7)[:-2]
DEN1["Origin Airport"] = "DEN"
DFW1 = pd.read_csv("Resources/CSV/American/DFW.csv", skiprows=7)[:-2]
DFW1["Origin Airport"] = "DFW"
JFK1 = pd.read_csv("Resources/CSV/American/JFK.csv", skiprows=7)[:-2]
JFK1["Origin Airport"] = "JFK"
LAS1 = pd.read_csv("Resources/CSV/American/LAS.csv", skiprows=7)[:-2]
LAS1["Origin Airport"] = "LAS"
LAX1 = pd.read_csv("Resources/CSV/American/LAX.csv", skiprows=7)[:-2]
LAX1["Origin Airport"] = "LAX"
MCO1 = pd.read_csv("Resources/CSV/American/MCO.csv", skiprows=7)[:-2]
MCO1["Origin Airport"] = "MCO"
ORD1 = pd.read_csv("Resources/CSV/American/ORD.csv", skiprows=7)[:-2]
ORD1["Origin Airport"] = "ORD"
PDX1 = pd.read_csv("Resources/CSV/American/PDX.csv", skiprows=7)[:-2]
PDX1["Origin Airport"] = "PDX"
SEA1 = pd.read_csv("Resources/CSV/American/SEA.csv", skiprows=7)[:-2]
SEA1["Origin Airport"] = "SEA"
SFO1 = pd.read_csv("Resources/CSV/American/SFO.csv", skiprows=7)[:-2]
SFO1["Origin Airport"] = "SFO"
PDX1
```

```python
PDX1 = filter(PDX1)
PDX1
```

```python
concat = pd.concat([ATL1, DEN1, DFW1, JFK1, LAS1, LAX1, MCO1, ORD1, PDX1, SEA1, SFO1])
concat['Carrier Code'] = "American Airlines"
filter(concat)
concat.set_index("Origin Airport")
```
