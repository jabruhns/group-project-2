```python
import os
import pandas as pd
```

```python
def filter(airport):
    airport = airport.rename(columns={'Destination Airport': 'Destination_Airport'})
    Destination_booleans = []
    for place in airport.Destination_Airport:
        if place == "SFO":
            Destination_booleans.append(True)
        elif place == "LAX":
            Destination_booleans.append(True)
        elif place == "DEN":
            Destination_booleans.append(True)
        elif place == "LAS":
            Destination_booleans.append(True)
        elif place == "PDX":
            Destination_booleans.append(True)
        elif place == "SEA":
            Destination_booleans.append(True)
        elif place == "DFW":
            Destination_booleans.append(True)
        elif place == "MCO":
            Destination_booleans.append(True)
        elif place == "ORD":
            Destination_booleans.append(True)
        elif place == "JFK":
            Destination_booleans.append(True)
        elif place == "ATL":
            Destination_booleans.append(True)
        else:
            Destination_booleans.append(False)
    is_long = pd.Series(Destination_booleans)
    airport = airport[is_long]
```
