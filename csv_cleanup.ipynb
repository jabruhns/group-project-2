```python
ATL2 = pd.read_csv("Resources/CSV/Alaska/ATL.csv", skiprows=7)[:-2]
ATL2["Origin Airport"] = "ATL"
ATL2 = filter(ATL2)
DEN2 = pd.read_csv("Resources/CSV/Alaska/DEN.csv", skiprows=7)[:-2]
DEN2["Origin Airport"] = "DEN"
DEN2 = filter(DEN2)
DFW2 = pd.read_csv("Resources/CSV/Alaska/DFW.csv", skiprows=7)[:-2]
DFW2["Origin Airport"] = "DFW"
DFW2 = filter(DFW2)
JFK2 = pd.read_csv("Resources/CSV/Alaska/JFK.csv", skiprows=7)[:-2]
JFK2["Origin Airport"] = "JFK"
JFK2 = filter(JFK2)
LAS2 = pd.read_csv("Resources/CSV/Alaska/LAS.csv", skiprows=7)[:-2]
LAS2["Origin Airport"] = "LAS"
LAS2 = filter(LAS2)
LAX2 = pd.read_csv("Resources/CSV/Alaska/LAX.csv", skiprows=7)[:-2]
LAX2["Origin Airport"] = "LAX"
MCO2 = pd.read_csv("Resources/CSV/Alaska/MCO.csv", skiprows=7)[:-2]
MCO2["Origin Airport"] = "MCO"
ORD2 = pd.read_csv("Resources/CSV/Alaska/ORD.csv", skiprows=7)[:-2]
ORD2["Origin Airport"] = "ORD"
PDX2 = pd.read_csv("Resources/CSV/Alaska/PDX.csv", skiprows=7)[:-2]
PDX2["Origin Airport"] = "PDX"
SEA2 = pd.read_csv("Resources/CSV/Alaska/SEA.csv", skiprows=7)[:-2]
SEA2["Origin Airport"] = "SEA"
SFO2 = pd.read_csv("Resources/CSV/Alaska/SFO.csv", skiprows=7)[:-2]
SFO2["Origin Airport"] = "SFO"
```

```python
concat2 = pd.concat([ATL2, DEN2, DFW2, JFK2, LAS2, LAX2, MCO2, ORD2, PDX2, SEA2, SFO2], sort=False)
concat2['Carrier Code'] = "Alaska Airlines"
concat2 = filter(concat2)
concat2
```

```python
AA_concat = pd.concat([concat, concat2])
AA_concat.set_index("Origin Airport")
```
